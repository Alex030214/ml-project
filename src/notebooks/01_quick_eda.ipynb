```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

ROOT   = Path('..')
DATA   = ROOT/'data'/'listings.csv'
FIGS   = ROOT/'figures'
RES    = ROOT/'results'
FIGS.mkdir(parents=True, exist_ok=True)
RES.mkdir(parents=True, exist_ok=True)

FIG_DPI = 300
plt.rcParams.update({'figure.figsize': (6,4), 'savefig.dpi': FIG_DPI})

df = pd.read_csv(DATA, low_memory=False)
print('Shape:', df.shape)
display(df.head(3))
display(df.dtypes.head(15))

if 'last_scraped' in df.columns:
    d = pd.to_datetime(df['last_scraped'], errors='coerce')
    print('Snapshot range:', d.min(), '→', d.max())

def clean_price(s):
    return (s.astype(str)
             .str.replace(r'[\$,]', '', regex=True)
             .str.replace(r'\.00$', '', regex=True)
             .replace('nan', np.nan)
             .astype(float))

if 'price' in df.columns:
    df['price'] = clean_price(df['price'])

na_rate = df.isna().mean().sort_values(ascending=False)
display(na_rate.head(20))
na_rate.to_csv(RES/'missing_rate.csv', header=['missing_fraction'])

if 'price' in df.columns:
    price = df['price'].dropna()
    if (price > 0).any():
        lo = max(price[price>0].min(), 10)
        hi = price.max()
        bins = np.logspace(np.log10(lo), np.log10(hi), 60)
        plt.figure()
        plt.hist(price[price>0], bins=bins)
        plt.xscale('log')
        plt.title('Price distribution (Boston Airbnb)')
        plt.xlabel('price (USD, log scale)')
        plt.ylabel('count')
        plt.tight_layout()
        plt.savefig(FIGS/'fig1_price_hist_log.png')
        plt.show()

if {'room_type', 'price'} <= set(df.columns):
    tmp = df[['room_type','price']].dropna()
    tmp = tmp[tmp['price'] > 0].copy()

    desired = ['Entire home/apt','Private room','Shared room']
    cats = [c for c in desired if c in tmp['room_type'].unique()]
    if not cats:
        cats = sorted(tmp['room_type'].unique())
    tmp['room_type'] = pd.Categorical(tmp['room_type'], categories=cats, ordered=True)

    fig, ax = plt.subplots(figsize=(6,4))
    tmp.boxplot(by='room_type', column='price', showfliers=False, ax=ax)
    ax.set_yscale('log')
    ax.set_title('Price by room_type (log y, no fliers + jitter sample)')
    ax.set_ylabel('price (USD, log scale)')
    ax.set_xlabel('room_type')
    fig.suptitle('')

    rng = np.random.default_rng(0)
    pos = {lab:i+1 for i,lab in enumerate(cats)}
    for lab in cats:
        s = tmp.loc[tmp['room_type']==lab, 'price']
        if len(s) == 0:
            continue
        s = s.sample(min(150, len(s)), random_state=0)
        x = rng.normal(loc=pos[lab], scale=0.06, size=len(s))
        ax.scatter(x, s.values, s=6, alpha=0.20, linewidths=0, zorder=2)

    plt.tight_layout()
    plt.savefig(FIGS/'fig2_roomtype_box_log_clean.png')
    plt.show()

plt.figure(figsize=(9,4.8))
ax = na_rate.head(12).plot(kind='bar')
plt.title('Top-12 missing rate')
plt.ylabel('fraction')
plt.xlabel('columns')
plt.xticks(rotation=65, ha='right', fontsize=10)
ax.tick_params(axis='x', length=0)
plt.tight_layout()
plt.savefig(FIGS/'fig3_missing_top.png')
plt.show()

keep_cols = [c for c in [
    'price','room_type','neighbourhood_cleansed','latitude','longitude',
    'minimum_nights','number_of_reviews','reviews_per_month',
    'availability_365','calculated_host_listings_count'
] if c in df.columns]

def col_kind(s): 
    return 'categorical' if s.dtype=='object' else 'continuous'

kinds = {c: col_kind(df[c]) for c in keep_cols}
type_tbl = pd.DataFrame.from_dict(kinds, orient='index', columns=['type'])
display(type_tbl)
type_tbl.to_csv(RES/'feature_types_selected.csv')

display(df[keep_cols].head())

if 'price' in df.columns:
    price = df['price'].dropna()
    summary = pd.Series({
        'count':  price.size,
        'mean':   round(price.mean(), 2) if price.size else np.nan,
        'median': round(price.median(), 2) if price.size else np.nan,
        'std':    round(price.std(), 2) if price.size else np.nan,
        'q25':    round(price.quantile(0.25), 2) if price.size else np.nan,
        'q75':    round(price.quantile(0.75), 2) if price.size else np.nan,
        'min':    round(price.min(), 2) if price.size else np.nan,
        'max':    round(price.max(), 2) if price.size else np.nan
    })
    display(summary.to_frame('price_stats'))
    summary.to_frame('price_stats').to_csv(RES/'price_summary.csv')

if {'neighbourhood_cleansed','price'} <= set(df.columns):
    top_med = (df[['neighbourhood_cleansed','price']]
               .dropna()
               .groupby('neighbourhood_cleansed')
               .filter(lambda x: len(x)>=20)
               .groupby('neighbourhood_cleansed')['price']
               .median()
               .sort_values(ascending=False)
               .head(10))
    ax = top_med.plot(kind='bar', figsize=(7,4), title='Top-10 neighbourhoods by median price')
    ax.set_ylabel('price (USD)')
    ax.set_xlabel('neighbourhood')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.savefig(FIGS/'fig4_neighbourhood_top10_median_price.png')
    plt.show()

if {'number_of_reviews','price'} <= set(df.columns):
    tmp = df[['number_of_reviews','price']].dropna()
    plt.figure(figsize=(6,4))
    plt.scatter(tmp['number_of_reviews'], tmp['price'], s=10, alpha=0.4)
    plt.title('Price vs number_of_reviews')
    plt.xlabel('number_of_reviews')
    plt.ylabel('price (USD)')
    plt.tight_layout()
    plt.savefig(FIGS/'fig5_price_vs_reviews.png')
    plt.show()

if {'latitude','longitude','price'} <= set(df.columns):
    m = df[['latitude','longitude','price']].dropna()
    if len(m) > 1000:
        m = m.sample(1000, random_state=0)
    plt.figure(figsize=(6,4))
    sc = plt.scatter(m['longitude'], m['latitude'], c=m['price'], s=8, alpha=0.6)
    plt.colorbar(sc, label='price (USD)')
    plt.title('Location scatter colored by price')
    plt.xlabel('longitude')
    plt.ylabel('latitude')
    plt.tight_layout()
    plt.savefig(FIGS/'fig6_geo_scatter_colored_by_price.png')
    plt.show()

num_cols_show = [c for c in ['price','latitude','longitude','minimum_nights','number_of_reviews',
                 'reviews_per_month','availability_365','calculated_host_listings_count'] if c in df.columns]
if num_cols_show:
    corr = df[num_cols_show].corr(numeric_only=True)
    plt.figure(figsize=(6,4))
    plt.imshow(corr, cmap='coolwarm', vmin=-1, vmax=1)
    plt.colorbar(label='corr')
    plt.xticks(range(len(num_cols_show)), num_cols_show, rotation=45, ha='right', fontsize=8)
    plt.yticks(range(len(num_cols_show)), num_cols_show, fontsize=8)
    plt.title('Correlation (numeric features)')
    plt.tight_layout()
    plt.savefig(FIGS/'fig7_corr_numeric.png')
    plt.show()

print("\nSaved to:")
print(" -", RES/'missing_rate.csv')
print(" -", RES/'feature_types_selected.csv')
print(" -", RES/'price_summary.csv')
print("Figures:")
for f in sorted(FIGS.glob('fig*.png')):
    print(" -", f.name)
```
